### MOST WINS IN IPL SEASON 2017,18,19

```python
temp=pd.DataFrame({"Winner": data1['winner']})
win_counts=temp.value_counts()
#print(win_counts)
label=[x[0] for x in win_counts.keys()]
bar, ax=plt.subplots(figsize=(20,12))    # 20 width 12 height
ax=plt.pie(x=win_counts,autopct='%.1f',labels=label)     # autopct==auto percentage
plt.show() 
```
